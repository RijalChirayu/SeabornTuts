## Advantages of Seaborn.
- Easy to use.
- Works well with pandas data structures.
- Built on top of matplotlib

The seaborn library was named after Samuel Norman Seaborn(SNS).

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
height = [62,64,69,74
          ,82,91,95,97]
weight = [120,136,145,148,
          156,161,186,205]

sns.scatterplot(x = height , y = weight)
plt.show()
```

### Count Plots
Count plots are used to display number of entries per category.

```python
gender = ['Female' , 'Male' , 'Female' , 
          'Female' , 'Male' , 'Male' , 'Female' , 'Female']
sns.countplot(x = gender)
plt.show()
```

Adding a third varibale with hue

```python
import pandas as pd
tips = sns.load_dataset('tips')
tips.head()
```

### WIthout hue

```python
sns.scatterplot(x="total_bill",
                y = "tip",
                data = tips)
plt.show()
```

```python
sns.scatterplot(x="total_bill",
                y = "tip",
                hue = "smoker" , 
                data = tips , 
                hue_order =['Yes' , 'No'])
plt.show()
```

### Specifying hue colors

```python
hue_colors = {"Yes":"r",
            "No": "b"
}
sns.scatterplot(x="total_bill" , y = "tip", 
                data = tips , hue = "smoker" , palette = hue_colors)
plt.show()
```

### Using hue with count plots

```python
new_colors = {"Male" : "#89B0AE" ,"Female" : "#FFD6BA" }
sns.countplot(x = "smoker" ,data = tips , hue = "sex" ,
palette = new_colors)
plt.show()
```

## Relational Plots and subplots

### Introduction to relplot()
It enables us to visualize relation between two quantitative variables using scatter plot or line plots.
- It lets us create subplots in a single figure. 

```python
sns.relplot(x = "total_bill" , y = "tip" , data = tips , 
            kind = "scatter" , col = "smoker"  , hue = "size")
plt.show()
```

```python
sns.relplot(x = "total_bill" , y = "tip" , data = tips , 
            kind = "scatter" , row = "smoker")
plt.show()
```

```python
sns.relplot(x = "total_bill" , y = "tip" , data = tips , 
            kind = "scatter" , col = "smoker" , row = "time" )
plt.show()
```

### Customizing subplots
Using size parameter

```python
sns.relplot(x = 'total_bill' , y = 'tip',
            data = tips , 
            kind="scatter", 
            size= "size" ,
            hue="size")
plt.show()
```

Using point style

```python
sns.relplot(x = 'total_bill' , y = 'tip',
            data = tips , 
            kind="scatter", 
            style= "smoker" ,
            hue="smoker")
plt.show()
```

Using point transparency

```python
sns.relplot(x = "total_bill" , 
            y = 'tip', data = tips ,
            kind="scatter" , 
            alpha = 0.4)
plt.show()
```

```python
mpg = sns.load_dataset('mpg'
                       )
sns.relplot(x="horsepower" ,
            y= "mpg" , 
            data = mpg , 
            size = 'cylinders' , kind = "scatter" , hue = "cylinders")
plt.show()
```

```python
sns.relplot(x="acceleration" ,
            y= "mpg" , 
            data = mpg , 
            style = 'origin' ,
            kind= "scatter" , 
            hue = "origin")
```

```python
mpg
```

```python
sns.relplot(x = "weight" , 
           y = "acceleration" , 
           data = mpg ,
           kind = "scatter",
           hue= "cylinders",
           col="origin"
           )
```

We can clearly see that there is negative correlation between weight, acceleration and number of cylinders. Most of the vehicles from USA are heavier than that of japan and europe.

### Line Plots in Seaborn

In seaborn there are two types of relational plots: 
- Scatter Plot
- Line Plot

```python
sns.relplot(x = "cylinders" , y="acceleration" , data = mpg , kind = "line")
```

Here the shaded region is confidence interval.

```python
sns.relplot(x = "cylinders" , y="acceleration" , data = mpg , kind = "line" , ci = None)
plt.show()
```

```python
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin" , dashes = False , markers = True)
```

### Creating box plot with seaborn

- A box plot shows the distribution of quantitative data.
- We can see median , spread , skewness and outliers
- Facilitates comparison between groups.

```python
sns.catplot(
    x = 'time' ,
    y = 'total_bill' , 
    data = tips , 
    kind = 'box' , 
    order = ['Dinner' , 'Lunch'] , 
    sym = ""  , #Will omit the outliers from the plot altogether
    whis = [5 , 95] # Here the lower whisk will be at 5th percential and upper whisk will be at 95th percentile
)
plt.show()
```

```python
tips[tips['time'] == 'Dinner'].describe()
```

### Point plots
- Points show mean of quantitative variable
-  Vertical lines show 95% confidence intervals


```python
sns.catplot(y = 'total_bill' ,
            x ='time' ,
            data = tips ,
            kind = 'point' ,
            hue = 'size'
        )
```

```python
# from numpy import median
sns.catplot(y = 'total_bill' ,
            x ='smoker' ,
            data = tips ,
            kind = 'point' ,
            # hue = 'size' ,
            # join = True ,
            capsize = 0 ,
            join = False # Removes the line joining two points
            
        )
```

### Changing style and palette

```python
sns.set_style('darkgrid')
sns.catplot(
    x = 'time' ,
    y = 'total_bill' , 
    data = tips , 
    kind = 'box' , 
    order = ['Dinner' , 'Lunch'] , 
    sym = ""  , #Will omit the outliers from the plot altogether
    whis = [5 , 95] # Here the lower whisk will be at 5th percential and upper whisk will be at 95th percentile
)
plt.show()
```

### Addin title and axis labels

### FacetGrid vs. AxesSubplot objects
Seaborn plots create two diggerent types of objects: 
- FacetGrid 
- AxesSubplot
<br>
A FacetGrid contains of one or more AxesSubplot. AxesSubplot is the type of plots which are single.



Adding title to a FacetGrid

```python
g = sns.catplot(x = "sex" , y = "tip" ,
                data = tips ,
                kind = "box")
g.fig.suptitle("Tips vs Gender" , y = 1.05)
```

Adding title to a axessubplot

```python
g = sns.boxplot(x = 'sex' ,
                y = 'tip',
                data = tips ,
                )

g.set_title("New Title" , y = 1.05)
```

Titles for subplots

```python
g = sns.catplot(x = 'sex' ,
                y = 'tip',
                data = tips ,
                col = 'smoker',
                kind = 'box')
g.fig.suptitle("Smoker?" , y = 1.05)
g.set_titles("This is Smoker" , "this is non smoker")
# g.set
```

```python
g = sns.catplot(x = 'sex' ,
                y = 'tip',
                data = tips ,
                kind = 'box'
                )
g.set(xlabel = "Gender" ,  #This works with both plot objects
      ylabel = "Tips given")
plt.xticks(rotation = 90) # We use matplotlib to rotate the ticksb
plt.show()
```
