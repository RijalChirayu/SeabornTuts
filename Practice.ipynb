```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
penguindf = sns.load_dataset('penguins')
```

```python
penguindf.dropna(inplace = True)
```

```python
sns.set_palette('RdBu')
g = sns.catplot(y = 'body_mass_g' , x = 'sex' , data = penguindf , kind= 'box' , hue = 'island')
g.fig.suptitle("Distribution of body mass of penguin"  , y = 1.04)
plt.show()
```
