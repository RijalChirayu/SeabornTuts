```python
import seaborn as sns
```

```python
titanicdf = sns.load_dataset('titanic')
```

```python
titanicdf
```

```python
sns.set_palette('PRGn')
```

### Creating a histogram

```python
sns.displot(titanicdf['age'] , kde = 'True' , bins = 10 )
plt.show()
```

```python
sns.displot(titanicdf['age'] , kind = 'kde' , rug = True ,fill = True )
plt.show()
```

```python
sns.displot(titanicdf['age'] , kind = 'ecdf'  )
plt.show()
```

### Regression Plots in seaborn

- The regplot() function is the basis of regression plots.
- The data x and y must be defined properly.

```python
sns.relplot(x = 'age' , 
            y = 'fare' , data = titanicdf ,
            kind = 'scatter')
```

```python
sns.regplot(x = 'Age' ,
            y = 'Weight' ,
            data = data )
```

```python
dfcorrel = pd.DataFrame({
    'Age' : list(titanicdf['age']) ,
    'Fare' : list(titanicdf['fare'])
    
})
```

```python
dfcorrel.corr()
```

Using dspine()

```python

sns.displot(titanicdf['age']  , bins = 10 )
sns.despine(left = False)
plt.show()
```

```python
fig , ax = plt.subplots()
sns.histplot(titanicdf['age'] , ax = ax)
ax.set(xlabel =' Tuition 2013-14' ,
       ylabel = 'Distribution' ,
       title ='TItle using axes')
```

### Categorical plots 
- Strip Plots
- Swarm plots

Strip plot

```python
sns.catplot(x = 'fare' ,
            y = 'sex' ,
            data = titanicdf ,
            kind = 'strip')
plt.show()
```

```python
sns.catplot(x = 'age' ,
            y = 'sex' ,
            data = titanicdf ,
            kind = 'swarm' )
plt.show()
```

```python
sns.catplot( x = 'class' ,
            y  = 'age' ,
            data = titanicdf ,
            kind = 'box' ,
           hue = 'sex')
plt.show()
```

Violin Plot

```python
sns.catplot( x = 'class' ,
            y  = 'age' ,
            data = titanicdf ,
            kind = 'violin' ,
           hue = 'sex')
plt.show()
```

The boxen plot is the hybrid between box plot and violin plot

```python
sns.catplot( x = 'class' ,
            y  = 'age' ,
            data = titanicdf ,
            kind = 'boxen' ,
           hue = 'sex')
plt.show()
```
